```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
#Reading the CSV
df=pd.read_csv(r"C:\Users\User\Downloads\data.csv",encoding= 'unicode_escape')
print(df)
#Replacing Unwanted values by NAN values
df['Description'] = df['Description'].str.replace(r'\W',"")
print(df)
#Dropping NAN Values from Description
df=df.dropna(subset = ['Description'])
#print(df)
#Dropping NAN values from CustomerID
df=df.dropna(subset = ['CustomerID'])
#print(df)
#Creating the new Column for predecting total Sales
df['Total Sales'] = df['Quantity'] *  df['UnitPrice']
print(df)
#Saving the CSV with different Name
#df.to_csv(r"C:\Users\User\Downloads\zyla2.csv")
#created another dataframe for furher cleaning the Data
df1=pd.read_csv(r"C:\Users\User\Downloads\zyla2.csv")
print(df1)
#converting Total Sales Values to Float
df1['Total Sales'] = df1['Total Sales'].apply(np.int64)
print(df1)
#According the given Problem we need to keep those products which gives best revenue,so we have neglected those products whose sales are less than 500 and taking those products whose sales are greater than 500
df1.drop(df1[df1['Total Sales'] < 500].index, inplace = True)
print(df1)
#Grouping the different products by aggregating total Sales.
df1=df1.groupby(['Description'], as_index=False)['Total Sales'].sum()
print(df1)
# Top 50 products which have high Sales
df1=df1.nlargest(50,['Total Sales'])
print(df1)
#visualization of top 50 products against there total Sales in bar chart
fig = plt.figure(figsize = (20, 10))
plt.bar(df1['Description'], df1['Total Sales'], color ='maroon') 
plt.xlabel("Products") 
plt.ylabel("Total No. Of Sales") 
plt.title("Total Sales Across Top 50 Products") 
plt.show() 
  








```
